```python
import os
import glob
import time
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np
import tensorflow as tf

# 파일 경로 설정
directory = '/content/drive/MyDrive/해커톤 사진 파일'  # your_folder_path를 실제 경로로 변경하세요

# 이전에 확인한 최신 파일 저장 변수
latest_file_checked = None

# TensorFlow 모델 로드 (필요한 경우 모델 로드 코드를 추가하세요)
# model = tf.saved_model.load('your_model_path')

def upload_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image

# 모델 예측 함수
def model_predict(image):
    converted_img = tf.image.convert_image_dtype(image, tf.uint8)
    result = model(converted_img)
    return result

# 바운딩 박스 그리기 함수
def draw_bounding_boxes(image, boxes, scores, classes, threshold=0.5):
    for box, score, cls in zip(boxes, scores, classes):
        if score > threshold:
            ymin, xmin, ymax, xmax = box
            (left, right, top, bottom) = (xmin * image.shape[1], xmax * image.shape[1],
                                          ymin * image.shape[0], ymax * image.shape[0])
            cv2.rectangle(image, (int(left), int(top)), (int(right), int(bottom)), (0, 255, 0), 2)
            cv2.putText(image, f'{int(cls)}: {score:.2f}', (int(left), int(top)-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    return image

# 이전 출력한 이미지 객체 저장 변수
previous_display = None

while True:
    # 이미지 파일 목록 가져오기 (예: .jpg, .png, .jpeg 파일)
    image_extensions = ['*.jpg', '*.jpeg', '*.png']
    files = []
    for ext in image_extensions:
        files.extend(glob.glob(os.path.join(directory, ext)))

    # 파일이 없는 경우 처리
    if not files:
        print("\r지정한 경로에 이미지 파일이 없습니다. 계속 대기 중...", end="")
    else:
        # 가장 최신 이미지 파일 찾기
        latest_file = max(files, key=os.path.getmtime)

        # 새로운 파일이 있으면 처리
        if latest_file != latest_file_checked:
            latest_file_checked = latest_file

            # 최신 이미지 파일 업로드
            image = upload_image(latest_file)

            # 모델 입력 크기에 맞게 이미지 전처리
            input_image = cv2.resize(image, (300, 300))  # 모델 입력 크기에 맞게 조정 (예: 300x300)
            input_image = input_image / 255.0  # 정규화
            input_image = np.expand_dims(input_image, axis=0)  # 차원 확장

            # 모델 예측
            predictions = model_predict(input_image)

            # 결과에서 박스, 점수, 클래스를 추출합니다.
            boxes = predictions['detection_boxes'].numpy()
            scores = predictions['detection_scores'].numpy()
            classes = predictions['detection_classes'].numpy()

            # 점수 기준으로 필터링 (예: 0.5 이상)
            threshold = 0.5
            filtered_boxes = boxes[0][scores[0] > threshold]
            filtered_scores = scores[0][scores[0] > threshold]
            filtered_classes = classes[0][scores[0] > threshold]

            # 바운딩 박스 그리기
            image_with_boxes = draw_bounding_boxes(image, filtered_boxes, filtered_scores, filtered_classes)

            # 이전 출력한 이미지 삭제
            plt.clf()

            # 결과 이미지 표시
            plt.figure(figsize=(10, 10))
            plt.imshow(image_with_boxes)
            plt.axis('off')
            plt.show()

    # 10초 대기 (필요에 따라 조정 가능)
    time.sleep(10)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def draw_points(image_path, bounding_boxes):
    # 이미지 읽기
    image_bbox = image_with_boxes
    height, width, _ = image_bbox.shape

    # 중앙 좌표를 저장할 리스트
    center_points = []

    for box in bounding_boxes:
        x, y, w, h = box
        # 중앙 좌표 계산
        center_x = int(x + w / 2)
        center_y = int(y + h / 2)
        center_points.append((center_x, center_y))

        # 중앙에 점 찍기
        cv2.circle(image, (center_x, center_y), 5, (0, 0, 255), -1)

    # 결과 이미지 저장 또는 표시
    cv2.imwrite('output_image.jpg', image)
    cv2.imshow('Image with Centers', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    return center_points

def analyze_density(center_points, grid_size, image_shape):
    height, width = image_shape
    grid_h = height // grid_size
    grid_w = width // grid_size

    density_grid = np.zeros((grid_h, grid_w))

    for point in center_points:
        x, y = point
        grid_x = x // grid_size
        grid_y = y // grid_size
        density_grid[grid_y, grid_x] += 1

    return density_grid

def plot_density(density_grid):
    plt.imshow(density_grid, cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.title('Density Heatmap')
    plt.show()

# 테스트 데이터
image_path = 'path_to_your_image.jpg'
bounding_boxes = [
    (50, 50, 100, 100),
    (150, 50, 100, 100),
    (300, 50, 100, 100),
    (450, 50, 100, 100)
]

# 이미지에 점 찍기
center_points = draw_points(image_path, bounding_boxes)

# 점의 밀집도 분석
grid_size = 50  # 그리드 크기
density_grid = analyze_density(center_points, grid_size, (600, 800))  # 이미지 크기 (height, width)

# 밀집도 시각화
plot_density(density_grid)
```

```python
import serial
import time

# 시리얼 포트 설정 (Windows의 경우 'COM3', Mac/Linux의 경우 '/dev/ttyUSB0' 또는 '/dev/ttyACM0' 등)
serial_port = 'COM6'  # 사용자의 환경에 맞게 수정
baud_rate = 9600  # 아두이노와 일치하는 시리얼 통신 속도
message = "Hello Arduino!"

try:
    ser = serial.Serial(serial_port, baud_rate, timeout=1)
    time.sleep(2)  # 시리얼 포트가 열릴 때까지 대기

    ser.write(message.encode())  # 데이터를 시리얼 포트로 전송
    print("Data sent to Arduino:", message)
    ser.close()  # 시리얼 포트 닫기
except serial.SerialException as e:
    print("Error opening serial port:", e)
```
